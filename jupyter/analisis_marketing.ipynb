## Descripción de campos

- **Campaign ID**: Identificador único de la campaña.
- **Company**: Nombre de la empresa que gestiona la campaña.
- **Campaign Type**: Tipo de campaña (por ejemplo, digital, impresa).
- **Target Audience**: Público objetivo al que está dirigida la campaña.
- **Duration**: Duración total de la campaña.
- **Channel Used**: Canal utilizado para la campaña (por ejemplo, redes sociales, televisión).
- **Conversion Rate**: Tasa de conversión lograda por la campaña.
- **Acquisition Cost**: Costo de adquisición por cliente.
- **ROI**: Retorno de la inversión (Return on Investment) de la campaña.
- **Location**: Ubicación geográfica donde se ejecuta la campaña.
- **Language**: Idioma utilizado en la campaña.
- **Clicks**: Número de clics recibidos durante la campaña.
- **Impressions**: Número de impresiones generadas.
- **Engagement Score**: Puntaje de interacción de la campaña.
- **Customer Segment**: Segmento de clientes al que se dirige la campaña.
- **Date**: Fecha de inicio de la campaña.
- **Total($)**: Total en dólares generado o invertido en la campaña.
